```python
import pandas as pd
path = "/content/drive/MyDrive/covid_death_dataset_csv.csv"
df = pd.read_csv(path)
```

```python
total_rows = df.shape[0]
```

```python
df = df[df.Province_State!='0']
dropped_rows = total_rows - df.shape[0]
print(f'Initial number of total rows: {total_rows}')
```

```python
total_null = (df['Deaths'].isnull().sum() + df['Confirmed'].isnull().sum())
```

```python
df['Deaths'] = df['Deaths'].fillna(0)
df['Confirmed'] = df['Confirmed'].fillna(0)
```

```python
print(f'Number of dropped rows: {dropped_rows}')
print(f'Number of total null values: {total_null}')
```
